# Q2: Create a one line chart using two arrays. The first array represents x and the second array represents y
| x     |     y    |
| ----- | -------- |
| 1    | 1       |
| 2    | 2       |
| 3    | 3       |
| 4    | 5       |
| 5    | 8       |
| 6    | 13      |
| 7    | 20      |

```python
# write your code here ^_^
x=np.array([1, 2, 3, 4, 5, 6, 7])
y=np.array([1, 2, 3, 5, 8, 13, 20])

plt.plot(x, y)

plt.show()
```

# Q3: Create a two lines chart using the same previous arrays. 
But this time:
- Mark each point with a circle.
- Use dashed lines.
- Set the size of the marker (circle) to size 10.
- Set the width of the line to 3.
- Add legends. 'Normal' for array x and 'Fast' for array y.

```python
# write your code here ^_^
plt.plot(x, color = 'blue', linestyle = '--', linewidth = 3, marker = 'o', markersize = 10,markerfacecolor = 'red')
plt.plot(y, color = 'red', linestyle = '--', linewidth = 3, marker = 'o', markersize = 10,markerfacecolor = 'blue')
plt.legend(['Normal','Fast'], shadow = True, loc = 0)
plt.show()
```

# Q4: Display multiple line charts in one row using the charts given below.
- Give each chart a title 
- Add a super title 'my data visulaization assignment'. 

```python
#super title = my data visulaization assignment

#title = First Chart
x_plot1 = np.array([1, 2, 3, 4, 5, 6, 7])
y_plot1 = np.array([1, 1, 2, 3, 5, 8, 13])

#title = Second Chart
x_plot2 = np.array([0, 1, 2, 3, 4, 5, 6])
y_plot2 = np.array([2, 4, 6, 8, 10, 12, 14])

#title = Third Chart
x_plot3 = np.array([0, 1, 3, 4])
y_plot3 = np.array([4, 6, 3, 4])
```

```python
# write your code here ^_^
fig, axes = plt.subplots(1, 3, figsize=(10, 6))
fig.suptitle('my data visulaization assignment', fontsize = 20)

#first chart (0 -> first column)
axes[0].set_title("First Chart")
axes[0].plot(x_plot1, y_plot1, color = 'blue')

#second chart (1 -> second column)
axes[1].set_title("Second Chart")
axes[1].plot(x_plot2, y_plot2, color = 'red')

#third chart (2 -> third column)
axes[2].set_title("Third Chart")
axes[2].plot(x_plot3, y_plot3, color = 'green')

plt.show()
```

# Q6: Create a scatter plot  using two arrays. The first array represents carAge and the second array represents carSpeed, set a specific color for each dot by using an array of different colors and add labels for the plot
| carAge | carSpeed |
| ----- | -------- |
| 5    | 99       |
| 7    | 86       |
| 8    |87      |
| 7    | 88       |
| 2    | 111       |
| 17    | 86       |
| 2    | 103      |
| 9    | 87       |
| 4    | 94       |
| 11    | 78      |

- Set a new color for each dot by using an array of different colors.
- Change the size of the dots to size 350.
- Add labels for x-axis and y-axis.
- Adjust the transparency of the dots with the alpha argument. let the alpha value be 0.4

```python
# write your code here ^_^
#defining arrays
carAge=np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11])
carSpeed=np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78])
dotColors=np.array(["#674EE4", "#AA1DEB", "#4DBEE3","#4DE3C2", "#69E34D", "#D4E34D", "#E3C74D", "#E39D4D", "#E37D4D", "#E3554D"])

#plotting using scatter graph
plt.scatter(carAge, carSpeed, s = 350, c= dotColors, marker = 'o', alpha=0.4) 
plt.xlabel('Car Age',fontsize = 20)
plt.ylabel('Car Speed',fontsize = 20)

plt.show()
```

# Q7: Create a pie chart using two arrays. The first array represents python_libraries and the second array represents your skills out of 10 at each library
| python_libraries |   
| ------------ | 
| NumPy        | 
| Pandas       | 
| Matplotlib   | 
| Seaborn      | 
| Plotly       | 
| Scikit-learn | 

- Set a new color for each wedge by using an array of different colors.
- Pull your favorite library wedge 0.3 from the center of the pie.
- Add a shadow to the pie chart

```python
# write your code here ^_^
#defining arrays
pythonLibraries=np.array(['NumPy', 'Pandas', 'Matplotlib', 'Seaborn', 'Plotly', 'Scikit-learn'])
myPythonSkills=np.array([5, 4, 3, 1, 1, 6])

#pie chart
plt.pie(myPythonSkills, labels=pythonLibraries, autopct='%1.2f%%',
        #sliced the "dotColors" array that has been used previously on the scatter plot
        colors =dotColors[3:],  startangle=180, shadow=True, radius=1, explode = [0,0.3,0,0,0,0]);
```

# Q8: Create a histogram with 250 values and 15 bars

```python
# write your code here ^_^

#getting fixed random values
np.random.seed(0)
#defining arrays
values=np.random.uniform(0, 250, 250)
#fetched the min and max values from "values" array to initialize the bars correctly and evenly
bars=np.linspace(values.min(), values.max(), 15)
#plotting the histogram
plt.hist(values, bins=bars);
```
